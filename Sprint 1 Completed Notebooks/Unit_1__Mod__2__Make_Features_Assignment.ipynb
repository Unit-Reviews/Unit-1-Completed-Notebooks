```python
# Drop columns that are made up of all NaN Values

df = df.drop(['url', 'member_id', 'desc', 'id'], axis=1)

# Check table to make sure colums were dropped

df.head()
```

```python
# Clean up the int-rate column
# Look at the first 10 values of the int_rate column

df['int_rate'].head(10)
```

```python
# Remove the leading space and percentage and cast the string value to a float
# This is a test to see if we can use this in a function on our dataframe

int_rate =' 13.56%'
float(int_rate.strip().strip('%'))
```

```python
# It works! Let's write a function that can do what we have written above to any
# string that is passed to it

def int_rate_to_float(value):
  return float(value.strip().strip('%'))
```

```python
# Test the function by calling it on our example

int_rate_to_float(int_rate)
```

```python
# Try it with a different possible rate, it works!

int_rate_to_float(' 14.57%')
```

```python
# Example of a Loop Method

int_rate_list = []

for cell in df['int_rate']:
  int_rate_list.append(int_rate_to_float(cell))

# Look at the first 10 values 

int_rate_list[0:10]
```

```python
# Now turn the list into a new dataframe column

df['int_rate_cleaned'] = pd.Series(int_rate_list)

# Check to see if the new column is there

df.head()
```

```python
# Apply the solution to every cell in the column using .apply to
# apply the function

df['int_rate'] = df['int_rate'].apply(int_rate_to_float)

# Check to make sure the changes were made to the original column

df.head()
```

## Cleaning up the Employee Title ['emp_title'] column


```python
# Explore the top 20 employment titles

df['emp_title'].value_counts(dropna=False).head(20)
```

```python
# Explore the amount of unique employment title there are currently

len(df['emp_title'].value_counts())
```

```python
# How many of the employment titles are null (NaN)?

df['emp_title'].isnull().sum()
```

```python
# Create some examples that represent the cases that we want to clean up

examples = ['owner', 'Supervisor', 'Project Manager', np.NaN]
```

```python
# Write a function to clean up these uses cases and increase uniformity
# If/else addresses NaN's as floats

def clean_title(title):
  if isinstance(title, str):
    return(title.strip().title())
  else:
    return "Unknown"

for title in examples:
  print(clean_title(title))
```

```python
# List comprehensions can combine function calls and for loops over lists
# into one succinct and fairly readable single line of code.

[clean_title(title) for title in examples]
```

```python
# The function works as expected. Time to apply it to the column

df['emp_title'] = df['emp_title'].apply(clean_title)
```
